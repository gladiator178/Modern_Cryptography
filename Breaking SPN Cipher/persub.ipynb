```python
# frequency analysis 
def freqana(s):
    k={}
    for i in range(0,len(s)):
            if (s[i] in k):
                k[s[i]] += 1
            else:
                k[s[i]] = 1
    alph="abcdefghijklmnopqrstuvwxyz"
    for i in alph:
        if(i not in k):
            k[i]=0
    return k
k=freqana(s)
myKeys = list(k.keys())
myKeys.sort()
keys = list(k.keys())
values = list(k.values())
sor = np.argsort(values)
ks = {keys[i]: values[i] for i in sor}
 
# ks = {i: k[i] for i in myKeys}
print(ks)
names = list(ks.keys())
values = list(ks.values())    


plt.bar(range(len(ks)), values, tick_label=names)
plt.show()
```

```python
d={}
for i in range(0,len(s)-1):
        if (s[i:i+2] in d):
            d[s[i:i+2]] += 1
        else:
            d[s[i:i+2]] = 1
        if (s[i:i+3] in d):
            d[s[i:i+3]] += 1
        else:
            d[s[i:i+3]] = 1
        if (s[i:i+4] in d):
            d[s[i:i+4]] += 1
        else:
            d[s[i:i+4]] = 1
# freqency of bigrams, trigrams, ...
fi={}
for ele in d:
    if(d[ele]!=1):
        fi[ele]=d[ele]
print(fi)
```

```python
# distance between bigrams, trigrams, ...
for i in range(0,len(s)-1):
    if(s[i:i+2] in fi):
        for j in range(i+1,len(s)-2):
            if(s[j:j+2]==s[i:i+2]):
                print(s[i:i+2],":",j-i)
                break
for i in range(0,len(s)-2):
    if(s[i:i+3] in fi):
        for j in range(i+1,len(s)-2):
            if(s[j:j+3]==s[i:i+3]):
                print(s[i:i+3],":",j-i)
                break
for i in range(0,len(s)-3):
    if(s[i:i+4] in fi):
        for j in range(i+1,len(s)-2):
            if(s[j:j+4]==s[i:i+4]):
                print(s[i:i+4],":",j-i)
                break
```
